```python
import random
import json
import os

users=[]
data= []
name = "";
email="";
id ="";


print("Cadastre o nome de usuário");
name = input()
print("Insira o e-mail")
email=input()
id = random.randint(1,100)
users=[id,name,email] # Changed from set to list
print("Usuário cadastrado",users)

data = {"Usuários": users}
print(data)

# Ensure the directory exists before writing the file
os.makedirs('src', exist_ok=True)

with open('src/data.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, indent=4, ensure_ascii=False)
print("Usuário cadastrado com sucesso")
```
