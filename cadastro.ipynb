```python
import json
import pandas as pd
from google.colab import files

carregar_json=[]
csv_arquivo ='arquivo_csv.csv'
baixar=''

with open('src/banco_de_dados/data.json' ,'r', encoding='utf-8') as f:
 carregar_json = json.load(f)
print(carregar_json)

# A variável carregar_json já contém o dicionário Python carregado do JSON.
# pd.read_json espera um caminho de arquivo ou uma string JSON.
# Para criar um DataFrame a partir do dicionário, use pd.DataFrame diretamente.
# Assumindo que a chave 'Usuários' contém os dados que você quer como uma única linha.
df = pd.DataFrame([carregar_json['Usuários']], columns=['id', 'nome', 'email'])
df.to_csv(csv_arquivo,index=False, encoding='utf-8')

print(f"arquivo gerado {csv_arquivo} com êxito")
baixar = input('Deseja baixar o arquivo? Escolha Y para sim N para não ')

if baixar == 'y' or baixar == 'Y' :
 files.download('arquivo_csv.csv')
else :
    print('Arquivo não baixado')
```

```python
from google.colab import files
import json
import pandas as pd
import io



carregar_arquivo = files.upload()

nome_arquivo = list(carregar_arquivo.keys())[0]

if nome_arquivo.endswith(".csv"):
  df = pd.read_csv(io.BytesIO(carregar_arquivo[nome_arquivo]))
elif nome_arquivo.endswith(".xlsx"):
  df = pd.read_excel(io.BytesIO(carregar_arquivo[nome_arquivo]))
else:
  print("formato nao suportado")
  df = None


if df is not None:
  json_arquivo =  df.to_json(orient="records", force_ascii=False, indent=4)

  with open('src/banco_de_dados/data.json', 'w', encoding='utf-8') as f:
    f.write(json_arquivo)
  print(json_arquivo)
```
